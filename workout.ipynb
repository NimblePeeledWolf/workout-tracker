```python
from pdf2image import convert_from_path
import pytesseract
import pandas as pd
import re

def extract_week_day_title(text):
    """Extracts 'Week X - Day Y' from text using regex."""
    match = re.search(r'Week \d+ - Day \d+', text)
    if match:
        return match.group(0)
    return None

def extract_text_with_ocr(pdf_path):
    """Extracts text from an image-based PDF using OCR."""
    # Convert PDF to images
    images = convert_from_path(pdf_path)
    extracted_data = []

    for page_num, image in enumerate(images, start=1):
        # Extract text using pytesseract
        text = pytesseract.image_to_string(image)

        # Extract title
        title = extract_week_day_title(text)
        if title:
            print(f"Title found on page {page_num}: {title}")

        # Example: Custom table processing (if tables are in text format)
        # You can customize this section to process table-like structures in text
        print(f"Text on Page {page_num}:\n{text[:500]}")  # Preview text (first 500 chars)

        # Save extracted text for further processing
        extracted_data.append({"Page": page_num, "Title": title, "Text": text})

    return extracted_data

# Example usage
pdf_path = "cxc-training-program.pdf"
ocr_data = extract_text_with_ocr(pdf_path)

# Save the raw text to a file for inspection
with open("ocr_extracted_text.txt", "w") as f:
    for page in ocr_data:
        f.write(f"Page {page['Page']}\nTitle: {page['Title']}\n{page['Text']}\n{'='*40}\n")  # Changed 'text' to 'page['Text']'

print("Text extraction complete. Check 'ocr_extracted_text.txt' for details.")
```
